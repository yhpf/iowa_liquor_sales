# Build your models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

```python
# Build models of total sales based on location, price per bottle (I choose Retail), total bottles sold. 
# You may find it useful to build models for each county, ZIP code, or city. (I choose City)
```

```python
# define X and y for test_varables
X = iowa_stores[['Bottles Sold', 'State Bottle Retail']]
X = sm.add_constant(X) # add constant
y = iowa_stores['Sale (Dollars)']
```

```python
test_varables = sm.OLS(y, X) # obs! look at the order of y and X, it's reversed
test_varables = test_varables.fit() #fit the model
predictions = test_varables.predict(X) # make a prediction based on X

print(test_varables.summary())
```

```python
# Above you can see that Bottles Sold and State Bottle Retailare good to use to get Total sales.
# Both R2 and adj_R2 = 0.725
# Total sale +1 bottle -> Total sale will increase 12.6651 USD.
# State Bottle Retailare +1 USD -> Total sale will increase 2.4770 USD.
```

```python
# define dep and indep for model 1
dep = iowa_stores['Sale (Dollars)']
indep = model1.drop(['Sale (Dollars)','ADAIR'], axis=1) # drop ADAIR to have something to bench against
indep = sm.add_constant(indep) # add constant
#(I just picked Adair, you could pick which ever city you like)
```

```python
model1a = sm.OLS(dep,indep)
model1a = model1a.fit() 
predictions = model1a.predict(indep)

print(model1a.summary())
```

```python
# Bottles Sold and State Bottle Retail are still OK
# SWISHER has the biggest beta coef. 74.3221 -> Makes SWISHER the best city to open a new store in. 
# LAWLER has the 2nd biggest beta coef. 60.7265 -> second best City to open a store in.
# WINDSORHEIGHTS has the third biggest beta coef. 49.1238 -> third best City to open a store in.
# bc. it will lead to higher total sales than if we choose other cities to open in.
# We will see if this is still true after a validation of the model.
```

```python
plt.scatter(predictions, dep, s=30, c='g', marker='+', zorder=10)
plt.xlabel("Predicted Values of Total Sales")
plt.ylabel("True Values of Total Sales")

plt.plot([0, np.max(dep)], [0, np.max(dep)])
print("MSE:", model1a.mse_model) # from documentation http://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.RegressionResults.html
# very high MSE
```

```python
# cross validation
# Choose number of fold between 3 and 7 (take an odd number). Fold = cv
```

```python
lr = LinearRegression()
scores = cross_val_score(lr, indep, dep, cv=5) #try with 5 folds
scores
```

```python
print(np.mean(scores))
print(np.var(scores))
```

```python
lr = LinearRegression()
scores = cross_val_score(lr, indep, dep, cv=3) #try with 3 folds
scores
```

```python
print(np.mean(scores))
print(np.var(scores))
```

```python
lr = LinearRegression()
scores = cross_val_score(lr, indep, dep, cv=7) #try with 7 folds
scores
```

```python
print(np.mean(scores))
print(np.var(scores))
```

```python
# The R2 mean is best at 3 folds I will go with 3 folds. The variance is the lowest at 3 folds as well.
```

```python
# Fit the model
lr.fit(indep, dep)
```

```python
predictions = cross_val_predict(lr, indep, dep, cv=3)
r2 = metrics.r2_score(y, predictions)
print(r2)
```

```python
# Over- or underfitted?
# test < training -> overfitt
# 0.717143027167 (test) < 0.719048299741 (training)
```
